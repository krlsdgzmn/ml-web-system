```python
import pandas as pd
```

```python
df = pd.read_csv("data/dataset.csv")
df.tail()
```

```python
df.order_week = df.order_week.str.replace("'", "")
```

```python
price_bin = {"0-499": 1, "500-999": 2, "1000-1499": 3, "1500-1999": 4, ">2000": 5}

discount_bin = {
    "0-24": 1,
    "25-49": 2,
    "50-74": 3,
    "75-99": 4,
    ">100": 5,
}

order_week = {
    "Week 1": 1,
    "Week 2": 2,
    "Week 3": 3,
    "Week 4": 4,
}

order_month = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12,
}

distance_from_seller_km_bin = {
    "0-199": 1,
    "200-399": 2,
    "400-599": 3,
    "600-799": 4,
    "800-999": 5,
    ">1000": 6,
}

order_status = {
    "Completed": 0,
    "Cancelled": 1,
}
```

```python
df["price"] = df.price.map(price_bin)
df["total_discount_php"] = df.total_discount_php.map(discount_bin)
df["order_week"] = df.order_week.map(order_week)
df["order_month"] = df.order_month.map(order_month)
df["distance_from_seller_km"] = df.distance_from_seller_km.map(
    distance_from_seller_km_bin
)
df["order_status"] = df.order_status.map(order_status)
df.tail()
```

```python
df.to_pickle("data/dataset_cleaned.pkl")
df.to_csv("data/dataset_cleaned.csv", index=False)
```
